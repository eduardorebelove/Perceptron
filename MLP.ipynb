```python
import numpy as np
```

```python
# Dados de entrada e saída
input_data = np.array([[0., 0.], [0., 1.], [1., 0.], [1., 1.]])
output_data = np.array([[0.], [1.], [1.], [0.]])
```

```python
# Parâmetros da rede neural
input_layer_size = 2
hidden_layer_size = 2
output_layer_size = 1

# Inicialização dos pesos
hidden_weights = np.random.rand(hidden_layer_size, input_layer_size + 1) * 2 - 1
output_weights = np.random.rand(output_layer_size, hidden_layer_size + 1) * 2 - 1
```

```python
# Parâmetros de treinamento
threshold = 1e-5
epochs = 100000
learning_rate = 1e-2

# Inicialização do erro e do número de épocas
square_error = threshold + 1
n_epochs = 0


while (square_error > threshold) and (n_epochs < epochs - 1):
    n_epochs += 1
    square_error = 0

    for i in range(len(input_data)):

        # Camada de entrada
        input_with_bias = np.insert(input_data[i], 0, 1)

        # Camada escondida
        hidden_layer = np.tanh(np.dot(hidden_weights, input_with_bias))

        # Adiciona bias à camada escondida
        hidden_layer_with_bias = np.insert(hidden_layer, 0, 1)

        # Camada de saída
        output_layer = np.tanh(np.dot(output_weights, hidden_layer_with_bias))

        # Calcula o erro médio quadrático
        square_error += np.sum(np.subtract(output_data[i], output_layer)**2 * 0.5)

        # Backpropagation para atualizar os pesos
        output_error = - (output_data[i] - output_layer) * (1 / 1.7159) * (1.7159 - output_layer) * (1.7159 + output_layer)
        hidden_error = output_error * output_weights[0, 1:] * (1 / 1.7159) * (1.7159 - hidden_layer) * (1.7159 + hidden_layer)

        # Atualiza pesos da camada escondida
        hidden_weights -= np.outer(hidden_error, input_with_bias) * learning_rate

        # Atualiza pesos da camada de saída
        output_weights -= np.outer(output_error, hidden_layer_with_bias) * learning_rate

    print("Iteration: %.f - MSE: %s " % (n_epochs, square_error))

    if square_error < threshold:
        print('***************************')
        print('***   Treino Concluido  ***')
        print('***************************')
    elif n_epochs >= epochs - 1:
        print("Encerrado por Estouro de Epoca")
```

```python
# Imprime os dados de entrada, saída e corretos do treino
print("Dados de Entrada:")
print(input_data)
print("Dados de Saída:")
print(output_data)

# Calcula a quantidade de erros e acertos
correct = 0
incorrect = 0

for i in range(len(input_data)):

    input_with_bias = np.insert(input_data[i], 0, 1)
    hidden_layer = np.tanh(np.dot(hidden_weights, input_with_bias))
    hidden_layer_with_bias = np.insert(hidden_layer, 0, 1)
    output_layer = np.tanh(np.dot(output_weights, hidden_layer_with_bias))

    if np.round(output_layer[0]) == output_data[i][0]:
        correct += 1
    else:
        incorrect += 1

print("Treino concluído na iteração:", n_epochs)
print("Quantidade de Acertos:", correct)
print("Quantidade de Erros:", incorrect)

# Imprime os valores onde a rede neural aprendeu
print("Valores onde a rede neural aprendeu:")

for i in range(len(input_data)):

    input_with_bias = np.insert(input_data[i], 0, 1)
    hidden_layer = np.tanh(np.dot(hidden_weights, input_with_bias))
    hidden_layer_with_bias = np.insert(hidden_layer, 0, 1)
    output_layer = np.tanh(np.dot(output_weights, hidden_layer_with_bias))

    if np.round(output_layer[0]) == output_data[i][0]:
        print("Entrada:", input_data[i], "Saída:", output_layer, "Correto!")
    else:
        print("Entrada:", input_data[i], "Saída:", output_layer, "Erro!")
```
